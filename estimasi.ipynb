```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Muat data penjualan
data_penjualan = pd.read_csv('data_penjualan.csv', names=['ID', 'Nama_Barang', 'Jumlah_Penjualan', 'Harga_Satuan'], header=None, skiprows=1)

# Tampilkan lima baris pertama dari data
print(data_penjualan.head())

# Analisis deskriptif
print(data_penjualan.describe())

# Visualisasi data
plt.figure(figsize=(10, 6))
sns.barplot(x='Nama_Barang', y='Jumlah_Penjualan', data=data_penjualan)
plt.title('Visualisasi Jumlah Penjualan Barang')
plt.xlabel('Nama Barang')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45)
plt.show()

# Korelasi antar variabel numerik
kolom_numerik = data_penjualan.select_dtypes(include=[np.number])
plt.figure(figsize=(10, 8))
sns.heatmap(kolom_numerik.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi Data Penjualan')
plt.show()
```
